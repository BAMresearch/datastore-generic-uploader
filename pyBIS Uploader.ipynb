```python
from pathlib import Path
import getpass             # module to prompt for a password
from pybis import Openbis

class FileImport:
    def setDataFilePath(self, filepath: Path):
        pass
    def setMetaData(self, **kwargs):
        pass
    def upload(self, **kwargs):
        pass

class MySAXSImport(FileImport):
    _filepath = None
    _previewPath = None
    _name = None
    def setDataFilePath(self, filepath: Path, previewPath: Path):
        print(f"Got file path: {filepath=}")
        self._filepath = filepath
        self._previewPath = previewPath
    def setMetaData(self, name):
        self._name = name # FIXME: validate this, use args package
    def upload(self, username, space_name):
        assert len(self._name)
        assert self._filepath.is_file()
        assert self._previewPath.is_file()
        o = Openbis('https://bam-openbis02.germanywestcentral.cloudapp.azure.com', verify_certificates=False)
        if not o.is_session_active():
            o.login('ibressle', getpass.getpass(), save_token=True)
        mat_proj = o.get_projects(space=space_name)
        proj_name = mat_proj.df.code[0] # grab the first project
        coll = o.get_collections(project=mat_proj.df.code[0]) # grab the first collection
        coll_name = coll.df.identifier[0]
        my_object = o.new_object(
            type = 'EXPERIMENTAL_STEP', # type for the object.
            #code = 'WS_OBJECT', #the code for the object. We can leave it empty and it will be generated automatically.
            space = space_name, #the space that we previously got
            project = proj_name, # this is the project that we previously created, saved in the variable "new_project"
            collection = coll_name, # this is the collection that we previously created, saved in the variable "new_collection"
            #props = {"$name": "testname", "bam_oe": "OE_6.5"}
            props = {"$name": self._name}
        )
        my_object.save()
        #print(f"file: {self._filepath}")
        my_dataset = o.new_dataset(
            type = 'RAW_DATA',
            collection = coll_name,
            object = my_object,
            files = [str(self._filepath)]
        )
        my_dataset.save()
        my_preview = o.new_dataset(
            type = 'ELN_PREVIEW',
            collection = coll_name,
            object = my_object,
            files = [str(self._previewPath)]
        )
        #print(f"Uploaded \"{my_object.props["$name"]}\" to {coll_name}")
        my_preview.save()
        return my_object, my_dataset
```
